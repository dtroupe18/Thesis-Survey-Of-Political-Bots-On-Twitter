```python
import os
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  
import datetime
```

```python
def load_csv_as_df(file_name, sub_directories, col_name=None):
    '''
    Load any csv as a pandas dataframe. Provide the filename, the subdirectories, and columns to read(if desired).
    '''
    # sub_directories = '/Data/'
    base_path = os.getcwd()
    full_path = base_path + sub_directories + file_name
    
    if col_name is not None:
        return pd.read_csv(full_path, usecols=[col_name])
    
    # print('Full Path: ', full_path)
    return pd.read_csv(full_path, header=0)


def load_full_master_list():
    # Load Master ID list before and After Kavanaugh Nomination
    kav_only_df = load_csv_as_df('Kav-MasterIDs.csv', '/Data/Master-Data/')
    almost_full_df = load_csv_as_df('MasterIDs-4.csv', '/Data/Master-Data/')
    df = pd.concat([kav_only_df, almost_full_df])
    df = df.sort_values('cap')
    df = df.drop_duplicates('user_id', keep='last')
    return df
   
    
def load_removed_users(master_df, error_id_df):
    # Loads user accounts that were removed by Twitter
    error_id_df = error_id_df['user_id'].astype('int64')
    id_list = list(error_id_df)
    removed_profiles = master_df[master_df['user_id'].isin(id_list)]
    return removed_profiles


def calculate_account_age_when_observed(df):
    # Uses the account creation time to calculate the accounts age in days
    df['tweet_time'] = pd.to_datetime(df['tweet_time'])
    df['user_created_at'] = pd.to_datetime(df['user_created_at'])
    ages = []
    today = datetime.datetime.now()

    for index, row in df.iterrows():
        age = days_between_dates(row['user_created_at'], row['tweet_time'])
        ages.append(age)
        
    df['age_in_days'] = ages
    return df


def calculate_days_since_seen(df):
    # Calculates the numbers of days that have passed since this bot was originally observed
    df['tweet_time'] = pd.to_datetime(df['tweet_time'])
    ages = []
    today = datetime.datetime.now()

    for index, row in df.iterrows():
        age = days_between_dates(row['tweet_time'], today)
        ages.append(age)
        
    df['days_since_observed'] = ages
    return df


def days_between_dates(date_one, date_two):
    """
    :param date_one: datetime.datetime
    :param date_two: datetime.datetime
        https://docs.python.org/3/library/datetime.html#datetime-objects
    :return: Int - number of days between those two dates
    """
    d1 = datetime.date(date_one.year, date_one.month, date_one.day)
    d2 = datetime.date(date_two.year, date_two.month, date_two.day)
    delta = d2 - d1
    
    if abs((d2 - d1).days) > 0:
        return abs(delta.days)
    else:
        # Account cannot be less than 1 day old
        return 1


def show_removed_stats(df, size=21418):
    removed_count = len(df)
    removed_percent = removed_count/size
    still_active_count = size - removed_count
    
    print('Out of ', size,' bots only ', removed_count, ' were removed.')
    print(removed_percent, '% of bot accounts were removed')
    print('Average CAP for a removed account ', df['cap'].mean())
    print('Average days since observation for a removed account ', df['days_since_observed'].mean())
    print('Average age in days for a removed account ', df['age_in_days'].mean())
    print('Average tweets for a removed account ', df['user_statuses_count'].mean())
    print('There are ', still_active_count, ' bots still active on Twitter')
    
    

def display_accounts_not_rechecked(recheck_df, active_df, removedIDs_df):
    # ids_to_recheck = recheck_df['user_id'].unique()
    still_active_ids = active_df['user_id'].unique()
    removed_ids = removedIDs_df['user_id'].unique()
    
    print('Starting with ', len(recheck_df), ' ids')
    recheck_df = recheck_df[~recheck_df['user_id'].isin(still_active_ids)]
    print('After removing still active IDs ', len(recheck_df))
    
    recheck_df = recheck_df[~recheck_df['user_id'].isin(removed_ids)]
    print('After removing removed IDs ', len(recheck_df))
    
    return recheck_df

def load_removed_ids():
    df = load_csv_as_df('RemovedIDs.csv', '/Data/Removed-Data/')
    df2 = load_csv_as_df('RemovedIDs-2.csv', '/Data/Removed-Data/')
    
    return pd.concat([df, df2])


def load_still_active():
    df = load_csv_as_df('StillActiveIDs.csv', '/Data/Removed-Data/')
    df2 = load_csv_as_df('StillActiveIDs-2.csv', '/Data/Removed-Data/')
    
    return pd.concat([df, df2])
    
    
def get_profiles_before(master_df, recheck_id_list):
    return master_df[master_df['user_id'].isin(recheck_id_list)]


```

```python
def add_old_cap_to_active_bots(active_df, before_recheck_df):
    old_caps = []
    cap_diffs = []
    for index, row in active_df.iterrows():
        uid = row['user_id']
        new_cap = row['cap']

        before_row = before_recheck_df.loc[before_recheck_df['user_id'] == uid]
        old_cap = before_row['cap'].iloc[0]
        cap_diff = new_cap - old_cap
        
        old_caps.append(old_cap)
        cap_diffs.append(cap_diff)

    active_df['old_cap'] = old_caps
    active_df['cap_diff'] = cap_diffs
    
    return active_df
```

```python
def create_bot_df(removed_df, still_bot):
    # combine removed and still active into one df with bool column for removed
    removed_id_list = removed_df['user_id'].tolist()
    
    bot_df = pd.concat([removed_df, still_bot])
    
    removed_statuses = []
    for index, row in bot_df.iterrows():
        user_id = row['user_id']
        
        removed_statuses.append(user_id in removed_id_list)
    
    bot_df['removed'] = removed_statuses
    return bot_df
```
